```python
def max_depth(G):
    d = 0
    for i in G.node:
        d = max(d,G.node[i]['depth'])
    return float(d)
```

```python
def odds_of_small_ris(x):
    from random import sample
    r = len(ris)
    n = len(x)
    m = median(x)
    sims = 0
    for i in range(10000):
        x = sample(range(r), n)
        if median(ris[x])<m:
            sims+=1
    return sims/10000.0
```

```python
def starter_join(G):
    d = datetime.datetime(2013,7,2)
    for i in G.node:
        d = min(d,G.node[i]['join_time'])
    return d
```

###Simply being early to sign up in the team doesn't lead to more shares

```python
time_diff = []
sigmas = []
for i in range(len(C)):    
    if networkx.number_of_nodes(C[i])>1:
        sj = starter_join(C[i])
        for j in C[i].nodes():
            c = C[i].node[j]['join_time']-sj
            time_diff.append(c.days * 86400 + c.seconds)
            #print time_diff[-1]
            #print C[i].node[j]['join_time']-sj
            sigmas.append(C[i].node[j]['sigma'])
            
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(time_diff, sigmas)
ylim(-10, ylim()[1])
xlim(-100000, xlim()[1])
ylabel("Number of Shares")
xlabel("Time Between A Recruit Registering and Their Team Starter Registering (seconds)")
title("Simply being early to sign up in the team doesn't lead to more shares")

m, b, r, p, stderr = linregress(time_diff, sigmas)
x = arange(xlim()[0], xlim()[1])
y = m*x+b
ax.plot(x,y)
text(.1, .5, "y = %f*x+%.1f, R^2=%.2f, p=%.3f"%(m,b,r**2, p), transform = ax.transAxes)
savefig("figure%i.png"%z, dpi=300, bbox_inches='tight')
z+=1
```
